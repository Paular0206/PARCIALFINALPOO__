```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
gdf_POBLACION_CUNDINAMARCA = gpd.read_file("POBLACION_CUNDINAMARCA.csv")
gdf_POBLACION_CUNDINAMARCA
```

```python
gdf_Municipios_de_Cundinamarca_2022 = gpd.read_file("Municipios_de_Cundinamarca_2022.geojson")
gdf_Municipios_de_Cundinamarca_2022
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

# Plotear los límites de los municipios
gdf_Municipios_de_Cundinamarca_2022.plot(ax=ax, color='lightgray', edgecolor='black')

# Plotear los datos de población
gdf_POBLACION_CUNDINAMARCA.plot(ax=ax, column='3', cmap='OrRd', legend=True)

# Añadir título
plt.title('Mapa de Población de los Municipios de Cundinamarca')

# Mostrar el mapa
plt.show()
```
